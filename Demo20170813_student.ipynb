```python
import requests
res = requests.get('http://www.appledaily.com.tw/realtimenews/section/new/')
print (res.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text,'lxml')
#soup.select('.rtddt a')
news = soup.select('.rtddt a')
#print(news[0])
print(news[0].select('h1')[0].text)
print(news[0].select('h2')[0].text)
print(news[0].select('time')[0].text)
print(news[0]['href'])
```

```python
domain = 'http://www.appledaily.com.tw'
newsarray = []
for record in soup.select('.rtddt a'):
    dic = {}
    dic['title'] = record.select('h1')[0].text
    dic['category'] = record.select('h2')[0].text
    dic['dt'] = record.select('time')[0].text
    dic['link'] = domain + record['href']
    #print(title,category,dt,link)
    newsarray.append(dic)
print(newsarray)    
#df.to_excl('tt')    
```

### 抓取內文資訊

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('http://www.appledaily.com.tw/realtimenews/article/international/20170813/1181417/%E7%94%98%E8%82%85%E9%A2%A8%E6%99%AF%E5%8D%80%E6%9A%B4%E9%9B%A8%E6%88%90%E6%B4%AA%E6%B5%81%E3%80%80%E8%BD%8E%E8%BB%8A%E8%A2%AB%E6%B2%96%E8%B5%B0296%E9%81%8A%E5%AE%A2%E5%8F%97%E5%9B%B0')
res.text
#soup = BeautifulSoup(res.text,'html.parser')
```

```python
def pageDetail(pageurl):
    dic={}
    res = requests.get(pageurl)
    soup = BeautifulSoup(res.text,'html.parser')
    dic['title'] = soup.select('#h1')[0].text
    dic['summary'] = soup.select('#summary')[0].text
    dic['dt'] = soup.select('.gggs time')[0].text
    dic['clicked'] = soup.select('.clicked')[0].text if len(soup.select('.clicked'))>0 else None
    return dic
url = 'http://www.appledaily.com.tw/realtimenews/article/international/20170813/1181417/%E7%94%98%E8%82%85%E9%A2%A8%E6%99%AF%E5%8D%80%E6%9A%B4%E9%9B%A8%E6%88%90%E6%B4%AA%E6%B5%81%E3%80%80%E8%BD%8E%E8%BB%8A%E8%A2%AB%E6%B2%96%E8%B5%B0296%E9%81%8A%E5%AE%A2%E5%8F%97%E5%9B%B0'
pageDetail(url)
```

```python
listurl = 'http://www.appledaily.com.tw/realtimenews/section/new/{}'
domain = 'http://www.appledaily.com.tw'
newsary = []
for page in range(0,2):
    section = requests.get(listurl.format(page))
    soup = BeautifulSoup(section.text,'html.parser')
    
    for news in soup.select('.rtddt a'):
        dic = {}
        dic['link'] = domain + news['href']
        
        dic['category'] = news.select('h2')[0].text
        dic.update(pageDetail(dic['link']))
        newsary.append(dic)
newsary    
```

```python
newspd = pandas.DataFrame(newsary)
newspd
#newspd.head()
newspd.to_excel('appledaily.xlsx')
```
